3.1. Створіть нову колонку в даних `Rating Category`, яка визначає категорію рейтингу (`Rating`): високий, середній чи низький за наступним правилом:
- якщо рейтинг вищий або рівний 8 - то він високий (`High`)
- якщо рейтинг від 6 (включно) до 8 (не включно) - то він середній (`Medium`)
- інакше - низький (`Low`).

```python
import numpy as np

conditions = [
    (df['Rating'] >= 8),
    (df['Rating'] >= 6) & (df['Rating'] < 8),
    (df['Rating'] < 6)
]

choices = ['High', 'Medium', 'Low']

df['Rating Category'] = np.select(conditions, choices, default='Unknown')
df.columns
```

```python
# or
def rating_cat(rating):
    if rating >= 8:
        return 'High'
    elif rating >= 6:
        return 'Medium'
    else:
        return 'Low'

df['Rating Category'] = df['Rating'].apply(rating_cat)
df.columns
```

```python
# or
def rating_cat(rating):
    return 'High' if rating >= 8 else 'Medium' if rating >= 6 else 'Low'

df['Rating Category'] = df['Rating'].map(rating_cat)
df.columns
```

```python
# or
bins = [0, 6, 8, 10]
labels = ['Low', 'Medium', 'High']

df['Rating Category'] = pd.cut(df['Rating'], bins=bins, labels=labels, right=False)
df.columns
```
